```python
import pandas as pd
import numpy as np
import joblib
import os
import json
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor 
from sklearn.metrics import accuracy_score, r2_score
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore', category=UserWarning)
```

```python
MODEL_DIR = "models"
os.makedirs(MODEL_DIR, exist_ok=True)
```

```python
df = pd.read_csv("data.csv")  
RANDOM_STATE = 42
```

```python
features_lists = {
    'Solubility': ['TPSA', 'HBD', 'HBA', 'RingCount', 'AromaticRings', 'AromaticRatio'],
    'Permeability': ['HBD', 'HBA', 'RotBonds', 'RingCount', 'AromaticRings', 'AromaticRatio'],
    'Stability': ['MW', 'TPSA', 'HBD', 'HBA', 'RingCount', 'AromaticRings', 'AromaticRatio'],
    'Drug_likeness': ['TPSA', 'RotBonds', 'RingCount', 'AromaticRings', 'AromaticRatio'],
    'Toxicity': ['MW', 'LogP', 'TPSA', 'HBD', 'HBA', 'RotBonds', 'RingCount', 'AromaticRings', 'AromaticRatio']
}
```

```python
metadata = {
    'models': {},
    'background_data_shap': {}
}
```

```python
name = 'Solubility'
features = features_lists[name]
X = df[features]
y = df['Solubility_mg_L'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=RANDOM_STATE)

scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = RandomForestRegressor(
    n_estimators=100, 
    max_depth=15, 
    random_state=RANDOM_STATE, 
    n_jobs=-1
)
model.fit(X_train_scaled, y_train)
r2 = r2_score(y_test, model.predict(X_test_scaled))
print(f"{name} R² Score: {r2:.4f}")

joblib.dump(model, os.path.join(MODEL_DIR, f'model_{name}.pkl'))
joblib.dump(scaler, os.path.join(MODEL_DIR, f'scaler_{name}.pkl'))
metadata['models'][name] = {'features': features, 'type': 'regression'}
metadata['background_data_shap'][name] = X_train_scaled.tolist()
```

```python
name = 'Permeability'
features = features_lists[name]
X = df[features]
y = df['Permeability']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=RANDOM_STATE)

scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression(random_state=RANDOM_STATE, max_iter=1000, class_weight='balanced')
model.fit(X_train_scaled, y_train)
acc = accuracy_score(y_test, model.predict(X_test_scaled))
print(f"{name} Accuracy: {acc:.4f}")

joblib.dump(model, os.path.join(MODEL_DIR, f'model_{name}.pkl'))
joblib.dump(scaler, os.path.join(MODEL_DIR, f'scaler_{name}.pkl'))
metadata['models'][name] = {'features': features, 'type': 'classification', 'classes': list(model.classes_)}
metadata['background_data_shap'][name] = X_train_scaled.tolist()
```
